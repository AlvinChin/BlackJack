```python
#Create hand of player with function to add cards and adjust values for ace
class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.cards2 = []
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
        self.card_val = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
        
    def add_card(self,card):
        for i in range(0,len(card)):
            if type(card) == tuple:
                self.cards.append(card)
                self.cards2.append(card[i][1])
            else:
                self.cards.append(card[i])
                self.cards2.append(card[i][1])
        for i in range(0,len(card)):
            if type(card) == tuple:
                card_value = self.card_val[card[1]]
            else:
                card_value = self.card_val[card[i][1]]
            self.value += card_value
            if type(card) == tuple:
                break
        return self.value
    
    def adjust_for_ace(self):
        if 'Ace' in self.cards2 and self.value<=31 and self.value>21:
            return self.value-10
        else:
            return self.value
```

```python
#class for chips to account for bets
class Chips:
    
    def __init__(self):
        self.total = 100  # This can be set to a default value or supplied by a user input
        self.bet = 0
        
    def win_bet(self,bet):
        return self.total + bet
    
    def lose_bet(self,bet):
        return self.total - bet
    
    def __str__(self):
        return f"Chips available: {self.total}"
    
    def chips_avail(self):
        return self.total
```
